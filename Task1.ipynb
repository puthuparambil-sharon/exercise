### **Markov**

We define a probability space $(\Omega, \mathcal{A}, P )$ and a measurable space $(S, \mathcal{B})$. The set $S$ represents the set of all possible states.
A Markov Chain is a stochastic process $(X_t)_{t\in\mathbb{N}}:(\Omega, \mathcal{A}, P) \rightarrow\{1,2, \ldots n\}$ with the property that
$$
P\left[X_{t+1} \in \mathcal{B} \mid X_{0}, . ., X_{t}\right]=P\left[X_{t+1} \in B \mid X_{t}\right]
$$
for $t\in\mathbb{N}$ and $B\in\mathcal{B}$.
The **idea** of a Markov chain is that the conditional probability of $X_{t+1}$ given the outcome of $X_1, ..., X_t$ depends only on the outcome of $X_t$, namely a memoryless process.

The first and easiest example of Markov chain in Life Insurance is with the state set $S=(*,\dagger)$ where $*$ denote the state *alive* and $\dagger$ the state *dead*. \\
We define
$$
p_{i j}(s, t):=P\left[X_{t}=j \mid X_{s}=i\right]
$$
the transition probability from state $i$ to state $j$. With this notation we can define 4 possible transition probabilities for the aforementioned example, namely
\begin{align}
& p_{**}(t) =1-q_{x+t} & & \text{probability to stay alive} \\
& p_{\dagger *}(t)= 0 & & \text{probability to resuscitate} \\
& p_{*\dagger}(t)=q_{x+t} & & \text{probability to die} \\
& p_{\dagger\dagger}(t)=1 & & \text{probability to stay dead}
\end{align}
One can similarly model a disability setup.

### **Chapman Kolmogorov**
For $0\leq r\leq s\leq t$ it holds
$$P(r,t)=P(r,s)\cdot P(s,t)$$
where $P(r,t)=(p_{ik}(r,t))_{(i,k)}$ is defined as above. We can rewrite the theorem component-wise:
$$p_{i k}(r, t)=\sum_{j \in S} p_{j k}(s, t) \cdot p_{i j}(r, s)$$

**Remark**:


**Proof**: For $s < t < u$ we have
\begin{align*}
    p_{ik}(s, u) &= P[X_u = k \mid X_s = i] \\
    &= P[X_u = k \cap \Omega \mid X_s = i] \\
    &= P\big[X_u = k \cap \bigcup_{j \in S} \{X_t = j\} \mid X_s = i\big] \\
    &= \sum_{j \in S} P[X_u = k, X_t = j \mid X_s = i]
\end{align*}

\begin{align*}
    P[X_u = k, X_t = j \mid X_s = i] &= \frac{P[X_u = k, X_t = j, X_s = i]}{P[X_s = i]} \\
    &= \frac{P[X_u = k, X_t = j , X_s = i]}{P[X_s = i]} \frac{P[X_t = j , X_s = i]}{P[X_t = j , X_s = i]}\\
    &= \frac{P[X_t = j , X_s = i]}{P[X_s = i]}\frac{P[X_u = k, X_t = j , X_s = i]}{P[X_t = j , X_s = i]} \\
    &= P[X_t = j \mid X_s = i] \cdot P[X_u = k \mid X_t = j, X_s = i] \\
    &= P[X_t = j \mid X_s = i] \cdot P[X_u = k \mid X_t = j] \\
    &= p_{ij}(s, t) \cdot p_{jk}(t, u)
\end{align*}

### **Cash-flows**

We define a Markov chain $(X_t)_{t\in\mathbb{N}}$ with state space $S$. For the times $t$ and $t+1$, for each transition $i\rightarrow j$ there are two different transactions:

1.   $a_i^{Pre}$ at the beginning of the time interval
2.   $a_{ij}^{Post}$ at the end of the time interval

So $a_i^{Pre}$ is the payment for being in the state $i$ at time $t$ and $a_{ij}^{Post}$ is the payment from going from state $i$ to state $j$ between $t$ and $t+1.

We define $I_i(t) = 1_{X_t = i}$, i.e. the indicator function that gives 1 if at time $t$ we are in state $i$. We get the annuity $a_i^{Pre}(t)$ at time $t$, so we get in total
$$
\sum_{i \in S} a_{i}^{pre}(t) \cdot I_i(t)
$$
Then we get a transaction for the death benefit of $a_{ij}^{Post}$ if we are in the state $i$ at time $t$ and at state $j$ at time $t+1$, so in total
$$
\sum_{(i,j) \in S^2} a_{ij}^{post}(t) \cdot I_i(t) \cdot I_j(t+1)
$$
Since we want $a_{ij}^{Post}$ to be compatible with $a_i^{Pre}$, we need to discount it with $v=\frac{1}{1+i}$ where $i$ is the interest rate. So we get the cash flow $A(t)$ and the discounted cash flow $\tilde{A}(t)$ given by
\begin{eqnarray}
A(t) &=& \sum_{i \in S} a_{i}^{pre}(t) \cdot I_i(t) + \sum_{(i,j) \in S^2} a_{ij}^{post}(t) \cdot I_i(t) \cdot I_j(t+1)\nonumber \\
\tilde{A}(t) &=& \sum_{i \in S} a_{i}^{pre}(t) \cdot I_i(t) + \sum_{(i,j) \in S^2} a_{ij}^{post}(t) \cdot I_i(t) \cdot I_j(t+1) \cdot v \nonumber
\end{eqnarray}

We can then compute the Mathematical Reserve as the conditional expectation of the PV of future Cash Flows, namely
$$
V_j(t)= \mathbb{E}\left[\sum_{\tau=0}^\infty v^{\tau} \tilde{A}(t+\tau) | \, X_t=j\right]
$$

Since for $A\in \mathcal{A}$ we have $\mathbb{E}[1_{A}]=P[A]$, then
\begin{align}
\mathbb{E}[I_i(t + \tau) | X_t = j]  &=  p_{ji}(t, t + \tau) \\
\mathbb{E}[I_i(t + \tau) \cdot I_k(t + \tau+1)| X_t = j] &= p_{ji}(t, t+\tau) \cdot p_{ik}(t+\tau,t+\tau+1)
\end{align}

Second equality (basically CK):
\begin{eqnarray}
\mathbb{E}[I_i(t + \tau) \cdot I_k(t + \tau+1)| X_t = j] & = & P[ X_{t+\tau + 1}= k,  X_{t+\tau}= i | X_t = j]\nonumber \\
 & = & \frac{P[X_{t+\tau + 1}= k,  X_{t+\tau}= i, X_t = j]}{P[X_t = j]}\nonumber \\
 & = & \frac{P[X_{t+\tau + 1}= k,  X_{t+\tau}= i, X_t = j]}{P[X_{t}= j,  X_{t+\tau}= i]} \cdot \frac{P[X_{t}= j,  X_{t+\tau}= i]}{P[X_t = j]}\nonumber \\
 & = & P[ X_{t+\tau+1}= k| X_t = j, X_{t+\tau}= i] \cdot P[ X_{t+\tau}= i| X_t = j]\nonumber \\
 & = & p_{ji}(t, t+\tau) \cdot p_{ik}(t+\tau,t+\tau+1)\nonumber
\end{eqnarray}

Altogether we get
\begin{align}
V_j(t)&= \mathbb{E}\left[\sum_{\tau=0}^\infty v^{\tau} \tilde{A}(t+\tau) | \, X_t=j\right] \\
& =\sum_{\tau=0}^\infty v^{\tau}\left(\sum_{i \in S} a_{i}^{pre}(t+\tau) \cdot \mathbb{E}[I_i(t+\tau)| \, X_t=j]  + \sum_{(i,k) \in S^2} a_{ik}^{post}(t+\tau) \cdot \mathbb{E}[I_i(t+\tau)\cdot I_k(t+\tau +1)| \, X_t=j]   \cdot v\right)\\
& = \sum_{\tau=0}^\infty v^{\tau} \left( \sum_{i \in S} a_{i}^{pre}(t+\tau) \cdot p_{ji}(t,t+\tau) + \sum_{(i,k) \in S^2} a_{ik}^{post}(t+\tau) \cdot p_{ji}(t,t+\tau) \cdot p_{ik}(t+\tau,t+\tau+1)\cdot v \right)
\end{align}

We then get the recursive formula (also called **Thiele Equation**)
$$
V_{i}(t)=a_{i}^{Pre}(t)+\displaystyle\sum_{j \in S} v_{i}(t) p_{i j}(t)\left\{a_{i j}^{P o s t}(t)+V_{j}(t+1)\right\}
$$

### **Decomposition of premium in risk and savings part, Technical Analysis**

Technical Analysis of an insurance policy: Comparing for a given period the incurred losses with the funds available to cover these losses.

How: Calculate the effective incurred losses and premium for the policies. The premium is then split into risk and savings premium.

First, one needs methods to calculate the effective incurred losses and the premiums for the policies. The latter will be split into a risk premium and a savings premium.

We first define Normal subsequent space by a function $\phi:S\to S, i\mapsto \phi(i)$ which assign to the state $i$ the subsequent state to it such that there is no payout according to the policy. It describes the transitions without incurring losses

We then define the Savings premium for state $i$ for time interval $(t,t+1]$ as
$$
\Pi_i^{(s)}(t) = v_t^i\cdot V_{\phi(i)}(t+1)-V_i(t)
$$
Equivalently, we have
$$
V_i(t) + \Pi_i^{(s)} = v_t^i\cdot V_{\phi(i)}(t+1)
$$
namely $\Pi_i^{(s)}$ is the amount of money we have to add to the reserve so that we have necessary funds in the reserve the next year in the normal subsequent state $\phi(i)$.

The technical pension debt (or regular cash flow) is then
$$
\Pi_i^{(tr)}(t) = a_i^{Pre}(t) + v_t^i \cdot a_{i\, \phi(i)}^{Post}(t)
$$

**For a policy with premiums:** $\Pi_i^{(tr)}(t)$ coincides with premium

Consider now $i$ and $j\neq \phi(i)$. The the value at risk $R_{ij}(t)$ is the loss incurred from going from $i$ to $j$ between $t$ and $t+1$, namely
$$
R_{ij}(t) = V_j(t+1)+a_{ij}^{Post}(t)-\big(V_{\phi(i)}(t+1)+a_{i\,\phi(i)}^{Post}(t)\big)
$$
The risk premium for going from $i$ to $j$ is then
$$
\Pi_{ij}^{(r)}(t) = p_{ij}(t)R_{ij}(t)
$$
and the total risk premium
$$
\Pi_i^{(r)}(t) = \sum_{j: j\neq\phi(i)} \Pi_{ij}^{(r)}(t)
$$

**Technical analysis:** Compare the loss which occurs in the year and the risk premium for the corresponding transition.

Loss quotient: $\frac{Loss(i\to j)}{\Pi_{}^{(r)}(t)}$

**Theorem:** $-\Pi_i^{(tr)}(t)=\Pi_i^{(r)}(t)+\Pi_i^{(s)}(t)$
